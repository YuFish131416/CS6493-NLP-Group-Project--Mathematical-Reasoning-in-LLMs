# 01 — Data Exploration
Explore the MATH-500, GSM8K, and AIME 2024 datasets.

```python
import sys
sys.path.insert(0, '..')

from src.data.loader import load_dataset, get_available_datasets, get_dataset_info
from src.data.preprocessor import preprocess_dataset

print('Available datasets:', get_available_datasets())
```

## Load MATH-500

```python
raw_math = load_dataset('math500', cache_dir='../data/')
math_problems = preprocess_dataset(raw_math, 'math500')
print(f'MATH-500: {len(math_problems)} problems')
print(f'Example:')
print(f'  Problem: {math_problems[0]["problem"][:200]}')
print(f'  Answer: {math_problems[0]["answer"]}')
print(f'  Level: {math_problems[0].get("level")}, Type: {math_problems[0].get("type")}')
```

```python
from collections import Counter
types = Counter(p.get('type') for p in math_problems if p.get('type'))
print('Problem types:', dict(types))
levels = Counter(p.get('level') for p in math_problems if p.get('level'))
print('Difficulty levels:', dict(sorted(levels.items())))
```
