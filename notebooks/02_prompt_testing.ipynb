## Answer Extraction Testing

```python
from src.evaluation.answer_extractor import extract_answer, extract_all_answers

test_responses = [
    'The answer is \boxed{5}',
    'After solving, we get #### 42',
    'Let me think step by step... Therefore, the answer is 7.',
    'The final answer is $x = 3$',
    'Step 1: ... Step 2: ... = 100',
]

for resp in test_responses:
    ans = extract_answer(resp)
    all_ans = extract_all_answers(resp)
    print(f'Response: {resp}')
    print(f'  Extracted: {ans}')
    print(f'  All candidates: {all_ans}')
    print()
```
